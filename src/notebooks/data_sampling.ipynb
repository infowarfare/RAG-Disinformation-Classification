```python
import pandas as pd

# Train data set
df = pd.read_csv("C:\\Users\\lucas\\Desktop\\RAG-Disinformation-Classification\\src\\pipeline_testing_data\\embedding_dataset.csv")

# 2. Definition der Prozentstufen
percentages = [0.10, 0.25, 0.50]

# 3. Erstellung der Subsets
subsets = {}
for p in percentages:
    # frac gibt den Anteil an (0.10 = 10%)
    # random_state sorgt für Reproduzierbarkeit
    subset_name = f"train_{int(p*100)}p"
    subsets[subset_name] = df.sample(frac=p, random_state=42)
    
    # Speichern als CSV
    subsets[subset_name].to_csv(f"C:\\Users\\lucas\\Desktop\\RAG-Disinformation-Classification\\src\\pipeline_testing_data\\{subset_name}.csv", index=False)
    print(f"Subset {subset_name} erstellt mit {len(subsets[subset_name])} Zeilen.")
```
